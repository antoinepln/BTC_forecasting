```python
import pandas as pd
```

```python
for x in range(1,11) :
    if x == 1 : 
        df = pd.read_json(f'perf_summary/5-1/dnn/2y/score_{x}.json')
    else :
        df = df + pd.read_json(f'perf_summary/5-1/dnn/2y/score_{x}.json')
df.index = ['acc', 'recall', 'speci', 'profit', 'min', 'max'] 
df.columns = [x for x in range(1,13)]
dnn_5_1_2y = df/10
```

```python
for x in range(1,11) :
    if x == 1 : 
        df = pd.read_json(f'perf_summary/5-1/dnn/3y/score_{x}.json')
    else :
        df = df + pd.read_json(f'perf_summary/5-1/dnn/3y/score_{x}.json')
df.index = ['acc', 'recall', 'speci', 'profit', 'min', 'max'] 
df.columns = [x for x in range(1,13)]
dnn_5_1_3y = df/10
```

```python
df_random = df = pd.read_json(f'perf_summary/baseline/score_random.json')
df_random.index = ['acc', 'recall', 'speci', 'profit', 'min', 'max'] 
df_random.columns = [x for x in range(1,13)]
```

```python
df_hold = df = pd.read_json(f'perf_summary/baseline/score_hold.json')
df_hold.index = ['acc', 'profit', 'min', 'max']
df_hold.columns = [x for x in range(1,13)]
```

```python
df_log = df = pd.read_json(f'perf_summary/baseline/log_model_1-1.json')
df_log.index = ['acc', 'recall', 'speci', 'profit', 'min', 'max'] 
df_log.columns = [x for x in range(1,13)]
```

```python
acc_compare = pd.concat((dnn_5_1_3y[0:1],dnn_5_1_2y[0:1],df_random[0:1], df_log[0:1]))
acc_compare.index = ['3y','2y','random', 'log' ]
```

```python
acc_compare
```

```python
recall_compare = pd.concat((dnn_5_1_3y[1:2],dnn_5_1_2y[1:2],df_random[1:2], df_log[1:2]))
recall_compare.index = ['3y','2y','random', 'log' ]
```

```python
min_compare = pd.concat((dnn_5_1_3y[5:6],dnn_5_1_2y[5:6],df_random[5:6], df_log[4:5], df_hold[3:4]))
min_compare.index = ['3y','2y','random', 'log','hold' ]
```

```python
min_compare
```

```python
tot_profit = {'3y': dnn_5_1_3y.loc['profit'].prod(), 
      '2y':dnn_5_1_2y.loc['profit'].prod(),
      'random' : df_random.loc['profit'].prod(),
      'hold': df_hold.loc['profit'].prod(),
      'log' : df_log.loc['profit'].prod()}
```
